```python
# Import the libraries
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.vgg16 import VGG16, preprocess_input
from tensorflow.keras.models import Model
from pathlib import Path
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import tempfile 
import os
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import splitfolders
```

```python
class FeatureExtractor:
    def __init__(self):
        # Use VGG-16 as the architecture and ImageNet for the weight
        base_model = VGG16(weights='imagenet')
        # Customize the model to return features from fully-connected layer
        self.model = Model(inputs=base_model.input, outputs=base_model.get_layer('fc1').output)
    def extract(self, img):
        # Resize the image
        img = img.resize((224, 224))
        # Convert the image color space
        img = img.convert('RGB')
        # Reformat the image
        x = image.img_to_array(img)
        x = np.expand_dims(x, axis=0)
        x = preprocess_input(x)
        # Extract Features
        feature = self.model.predict(x)[0]
        return feature / np.linalg.norm(feature)
```

### We are using vgg16 with imagenet weights and we'll do basic image processing before using it in the model. Then we'll get the image embedding through vgg16 using keras 'Model'. 

```python
embeddings=[]
path="C:\\Users\\Dell\\Documents\\projects\\image search engine\\oxford-iiit-pet\\images.tar\\images\\images\\"
fe=FeatureExtractor()
for img_path in os.listdir(path):
    print(img_path)
    # Extract Features
    img_path=os.path.join(path,img_path)
    feature =fe.extract(img=Image.open(img_path))
    # Save the Numpy array (.npy) on designated path
    #feature_path ="C:\Users\Dell\Documents\projects\image search engine\oxford-iiit-pet\images.tar\images\images\emb.npy"
    #np.save(r"C:\\Users\\Dell\\Documents\\projects\\image search engine\\oxford-iiit-pet\\images.tar\\images\\emb.npy", feature)
    embeddings.append(feature)

    
#     C:\Users\Dell\Documents\projects\image search engine\oxford-iiit-pet\images.tar\images\images
```

### We find embeddings to all the images and then store it in .npy format

```python
paths=[]
for img_path in os.listdir(path):
    paths.append(img_path)
```

### Storing the embeddings and retreiving it.

```python

# Insert the image query
img = Image.open(r"C:\Users\Dell\Documents\projects\image search engine\oxford-iiit-pet\images.tar\images\images\american_pit_bull_terrier_97.jpg")
pathn="C:\\Users\\Dell\\Documents\\projects\\image search engine\\oxford-iiit-pet\\images.tar\\images\\images\\"
# Extract its features
query = fe.extract(img)
# Calculate the similarity (distance) between images
dists = np.linalg.norm(embeddings - query, axis=1)
# Extract 30 images that have lowest distance
ids = np.argsort(dists)[:30]
scores = [(dists[id], paths[id]) for id in ids]
# Visualize the result
axes=[]
fig=plt.figure(figsize=(8,8))
for a in range(5*6):
    score = scores[a]
    axes.append(fig.add_subplot(5, 6, a+1))
    subplot_title=str(score[0])
    axes[-1].set_title(subplot_title)  
    plt.axis('off')
    new=os.path.join(pathn,score[1])
    plt.imshow(Image.open(new))
fig.tight_layout()
plt.show()
print(scores)
```

### Steps involved:
#### 1. We will find the image embedding for the query image.
#### 2. Then we'll find the distance (norm) of than image from all other image embeddings.
#### 3. We'll get ids of 30 images that have similar norms using argsort.
#### 4. We'll output the results using matplotlib subplot.
